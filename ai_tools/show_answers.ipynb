```python
import pickle
import gzip

with gzip.open('answers.pkl.gz', 'rb') as f:
    answers = pickle.load(f)

# Find the last question with a real answer
last_real_q = None
last_real_a = None
real_answer_count = 0

for q, a in answers.items():
    if a is not None and str(a).strip() != "" and str(a).strip().lower() != "none":
        last_real_q = q
        last_real_a = a
        real_answer_count += 1

if last_real_q is not None:
    print("Last question with a real answer:")
    print(last_real_q)
    print("\nAnswer:")
    print(last_real_a)
    print(f"\nTotal number of real answers: {real_answer_count}")
else:
    print("No real answers found.")
```
